```python
import nltk
sentence = "AI is the simulation of human intelligence by machine.it includes expects system NLP speech recognition and machine vision"
tokens = nltk.word_tokenize(sentence)
bigrams = list(nltk.bigrams(tokens))
for bigrams in bigrams:
    print(bigrams)
```

```python
import nltk
sentence = "AI is the simulation of human intelligence by machine.it includes expects system NLP speech recognition and machine vision"
tokens = nltk.word_tokenize(sentence)
trigrams = list(nltk.trigrams(tokens))
for trigrams in trigrams:
    print(trigrams)
```
